```c++
#include<fstream>
#include<iostream>
#include<math.h>
#include"map.h"
using namespace std;

// ERRO 1: 100 cm não era suficiente para guardar o mapa pois estourava o indice de coordenadas onde estão os obstáculos
// MELHORIA 1: usar uma constante ajuda nas modificações futuras e no calculo da translação das coordenadas

#define TAM 110

typedef struct {
    
    bool mapa[TAM][TAM];

}tMatriz;

typedef struct {

    float x;
    float y;
    float rot;
    float sensor[8];
    

}tCoordenadas
```

```c++
tMatriz inicializar_mapa(){
     
    tMatriz matriz;
    for(int i = 0; i < TAM;i++)
        for(int j = 0; j < TAM; j++)
            matriz.mapa[i][j] = 1;
    return matriz;
}
```

```c++
tCoordenadas ler_coordenadas(ifstream &file){
    
    tCoordenadas coords;
    file >> coords.x >> coords.y >> coords.rot;
        
        for(int i = 0; i < 8; i++)
            
            file >> coords.sensor[i];
            
         
    return coords;
    
}
```

```c++
tMatriz gerar_mapa(string nome_do_arquivo){
    
    tMatriz matriz = inicializar_mapa();
    ifstream file;
    file.open(nome_do_arquivo);
    //MELHORIA 2: desnecessária essa variável
    //float valor_do_sensor = 8.5;
    //ERRO 2: faltou declarar os angulos dos sensores
    float angulo[8] = {1.27, 0.77, 0.0, 5.21, 4.21, 3.14159, 2.37, 1.87};
    
    while(!file.eof()){
        tCoordenadas cds = ler_coordenadas(file);
        //ERRO 3: a conversão para cm e a translação devem ser feitas após o cálculo dos obstáculos
        //tCoordenadas cds2 = aproximar_e_converter_para_cm(cds);
        //tCoordenadas cds3 = enquadrar(cds2);
        
        
        for(int j =0; j < 7; j++){
            if(cds.sensor[j] < 0.085){
                //MELHORIA 2: desnecessário armazenar em outra variável
                //valor_do_sensor = cds3.sensor[j];
                     
                //ERRO 4: a formula correta necessita do angulo de cada sensor j conforme descrito no passo 4
                float x = cds.x + cds.sensor[j]*cos(cds.rot + angulo[j]);
                //ERRO 5: a formula correta utiliza o seno para projeção no eixo y
                float y = cds.y - cds.sensor[j]*sin(cds.rot + angulo[j]);

                //MELHORIA 3: translação e conversão para cm
                int xx = (int)(100 * x) + TAM/2;
                int yy = (int)(100 * y) + TAM/2;
                 
                matriz.mapa[xx][yy] = 0;
            }
            //MELHORIA 4: desnecessária essa lógica
            //else
            //    valor_do_sensor = 8.5;
        }
        //ERRO 6: O código abaixo deveria ser executado apenas quando encontra obstáculo. Ou seja, deveria ficar dentro do if acima
        //int x = cds3.x + valor_do_sensor*cos(cds3.rot + 1.27);
        //int y = cds3.y - valor_do_sensor*cos(cds3.rot + 0.77);    
        //matriz.mapa[x][y] = 0;
        
    }
   
    file.close();
    return matriz;
}
```
